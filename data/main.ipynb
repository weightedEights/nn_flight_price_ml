## Preprocessing

```python
# drop the unnamed and flight number columns
df = df.drop("Unnamed: 0", axis=1)
df = df.drop("flight", axis=1)
```

```python
# map class column to binary, 1 if business else 0
df["class"] = df["class"].apply(lambda x: 1 if x == "Business" else 0)
```

```python
# map stops column to ordinal with 0, 1, 2 as zero, one, two or more
df.stops = pd.factorize(df.stops)[0]
```
